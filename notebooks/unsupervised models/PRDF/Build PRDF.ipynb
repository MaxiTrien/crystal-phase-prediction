```python
dotants = pd.read_pickle(r'C:\Python\Projects\crystal-phase-prediction\data\pkl_files\probe_data.pkl')
dot = dotants['dopant']
dot
```

```python
dot = np.unique(dot)
dot = dot.tolist()
dot.remove('Hf') # or Zr
print(dot)
```

```python
from matminer.featurizers.structure import PartialRadialDistributionFunction

prdf = PartialRadialDistributionFunction(cutoff = 15, bin_size = 0.5, include_elems=['Hf', 'O'], exclude_elems=dot)
df = prdf.fit_featurize_dataframe(df, 'structure')
df
```
