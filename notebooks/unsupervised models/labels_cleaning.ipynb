# Labels cleaning

We got the labels from a csv file and now have to clean them up for later use of prediction. We will continue like the following: 
1. Delete the unimportent features
2. Build identifier columns for later reference
3. Change labels to phase columns

```python
import numpy as np
import pandas as pd
#pd.set_option('display.max_columns', None)
#pd.set_option('display.max_rows', None)
```

```python
abs_path = r'C:\Python\Projects\crystal-phase-prediction\crystal_data\overview.csv'
df = pd.read_csv(abs_path)
df
```

```python
# A lot of features that we dont need right now
list(df.columns)
```

```python
# drop columns that we dont need 
clean_df = df[['dopant', 'host, type', 'nominal concentration [f.u.%]', 'abc [sc]', 'phase', 'crit. overall']].copy()
clean_df
```

```python
# replacing the phase numbers with actual names
clean_df['phase'].replace({14:'m', 29:'p-o', 61:'o', 137:'t'}, inplace=True)
# print the number of phase changes in the overall dataset
print(clean_df['crit. overall'].value_counts())
clean_df
```

```python
# all columns except for crit. overall into on string collum as id
clean_df.dtypes
```

```python
clean_df['ID1'] = clean_df['dopant'].astype(str) + '_' + clean_df['host, type'].astype(str) + '_' + clean_df['nominal concentration [f.u.%]'].astype(str) + '_' + clean_df['abc [sc]'].astype(str) + '_' + clean_df['phase']
```

```python
clean_df
```

```python
# drop all the used features
clean_df = clean_df[['ID1', 'crit. overall']].copy()
clean_df
```

```python
# clean the , in the ID column
clean_df['ID1'] = clean_df['ID1'].str.replace(', ', '_')
clean_df
```

```python
# change the column names 
clean_df = clean_df.rename(columns={'ID1':'ID', 'crit. overall':'labels'})
clean_df
```

```python
# sort the values in the same order, then the data is loaded from a windows directory
clean_df = clean_df.sort_values('ID')
clean_df.head(25)
```

```python
# reorder the index column
clean_df = clean_df.reset_index(drop=True)
clean_df
```

```python
clean_df['new_labels'] = clean_df['ID'].str.extract(r'([^_]+$)')

clean_df['new_labels'] = clean_df['new_labels'].replace({'m': '0', 'p-o':'1', 'o':'2', 't':'3'})
clean_df
```

```python
# change the unknown phase in labels = False to 4 in new_labels
clean_df.loc[clean_df.labels == False, ['new_labels']] = 4
clean_df = clean_df.drop(columns = ['labels'])
clean_df
```

```python
# path to save labels df 
import pathlib as path 
t_path = path.WindowsPath('C:\Python\Projects\Thesis\data_labels\labels_all.pkl')
clean_df.to_pickle(t_path)
```
