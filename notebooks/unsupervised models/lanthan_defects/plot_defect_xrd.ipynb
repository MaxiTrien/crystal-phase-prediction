```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_pickle(r'C:\Python\Projects\crystal-phase-prediction\data\pkl_files\descriptors\xrd_La_defects5_65_m.pkl')
df
```

```python
labels = pd.read_pickle(r'C:\Python\Projects\crystal-phase-prediction\data\data_labels\La_defects_labels_m.pkl')
labels
```

```python
df_joined = pd.concat([df, labels], axis=1)
df_joined
```

```python
df_2La = df_joined[df_joined['labels'].str.contains('2La',case=False)]
df_La2V =  df_joined[df_joined['labels'].str.contains('La2V',case=False)]
df_inter =  df_joined[df_joined['labels'].str.contains('inter',case=False)]
df_La2 =  df_joined[df_joined['labels'].str.contains('La2',case=False)]

X_2La = df_2La.iloc[:, 2:-1].values
X_La2V = df_La2V.iloc[:, 2:-1].values
X_inter =  df_inter.iloc[:, 2:-1].values
X_La2 = df_La2.iloc[:, 2:-1].values
```

```python
theta2 = np.array(range(5, 66))
```

```python
fig=plt.figure(figsize=(13, 7))
rang = [X_2La.shape[0], X_La2V.shape[0], X_inter.shape[0], X_La2.shape[0]]
df_arr = [X_2La, X_La2V, X_inter, X_La2]
for ran, X in zip(rang, df_arr):
    for i in range(ran):
        plt.plot(theta2, X[i, :], alpha= 0.4)
    
plt.xlim([theta2.min(), theta2.max()])
plt.xlabel('$2theta)$')
plt.ylabel('Diffracted intensity (A.u.)')
plt.title('Crystal diffraction HfO2 defects')
```
