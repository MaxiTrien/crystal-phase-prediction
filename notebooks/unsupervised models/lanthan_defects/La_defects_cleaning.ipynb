```python
# for defects detection 
df_Hf14La2O32 = df[df['name'].str.contains('Hf14La2O32',case=False)]
df_Hf30La2O64 =  df[df['name'].str.contains('Hf30La2O64',case=False)]
df_Hf15LaO31 =  df[df['name'].str.contains('Hf15LaO31',case=False)]
df_Hf31LaO63 =  df[df['name'].str.contains('Hf31LaO63',case=False)]
df_Hf30La2O63 =  df[df['name'].str.contains('Hf30La2O63',case=False)]
df_La_HfO2_cat =  df[df['name'].str.contains('La_HfO2_cat',case=False)]
df_La_HfO2_inter =  df[df['name'].str.contains('La_HfO2_inter',case=False)]


df_Hf14La2O32['labels'] = df_Hf14La2O32['name'].str.extract(r'(Hf14La2O32)') # zwei La
df_Hf30La2O64['labels'] = df_Hf30La2O64['name'].str.extract(r'(Hf30La2O64)') 
df_La_HfO2_cat['labels'] = df_La_HfO2_cat['name'].str.extract(r'(cat)') # zwei La

df_Hf15LaO31['labels'] = df_Hf15LaO31['name'].str.extract(r'(Hf15LaO31)') # LaV
df_Hf31LaO63['labels'] = df_Hf31LaO63['name'].str.extract(r'(Hf31LaO63)')

df_Hf30La2O63['labels'] = df_Hf30La2O63['name'].str.extract(r'(Hf30La2O63)') # La2V

df_La_HfO2_inter['labels'] = df_La_HfO2_inter['name'].str.extract(r'(inter)') # inter


df = pd.concat([df_Hf14La2O32, df_Hf30La2O64, df_La_HfO2_cat, df_Hf15LaO31, df_Hf31LaO63, df_Hf30La2O63, df_La_HfO2_inter], axis=0)
df
```

```python
df['labels'].replace({'Hf14La2O32': '2La', 'Hf30La2O64': '2La', 'cat': '2La', 'Hf15LaO31': 'LaV', 'Hf31LaO63':'LaV',
                     'Hf30La2O63': 'La2V'},
                          inplace=True)
df
```
